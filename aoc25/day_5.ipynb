```python
# def function to read txt file into lines 
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
data = read_file_into_lines("C:/Users/gabia/Desktop/python_homework_gabi/data/advent_of_code_day_5.txt")
```

```python
def fresh_ranges(s):
    a, b = s.split('-')
    a, b = int(a), int(b)
    return a, b
```
