```python
# def function to read txt file into lines 
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
# use function for data set
data = read_file_into_lines("C:/Users/gabia/Desktop/python_homework_gabi/data/advent_of_code_day_5.txt")
```

```python
# get just ranges from dataset
just_ranges = []
for i in data:
    if i == "":
        break
    #if i == data[data.index("")-1]:
    just_ranges.append(i)

# get just IDs from dataset:

IDs = []
for i in data[data.index("")+1:]:
        IDs.append(int(i))
    

def fresh_ranges(s):
    a, b = s.split('-')
    a, b = int(a), int(b)
    return a, b
```

```python
# function to put ranges into start & stop
def find_start_and_stop(testranges,i):
    #for i in range(len(testranges)):
    range1 = testranges[i]
    start = range1[:range1.index("-")]
    stop = range1[range1.index("-")+1:]
    return start,stop

for index in range(len(just_ranges)):
    starts = find_start_and_stop(just_ranges,index)
 
```

```python
def count_fresh_products(testIDs, just_ranges):
    fresh = []
    for id in testIDs:
        for index in range(len(just_ranges)):
            startend = find_start_and_stop(just_ranges,index)
            key = int(startend[0])
            value = int(startend[1])
            if id >= key and id <= value:
                fresh.append(id)
    fresh = set(fresh)
    return len(fresh)
```
