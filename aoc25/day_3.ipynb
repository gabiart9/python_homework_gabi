```python
# def function to read txt file into lines 
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
data = read_file_into_lines("C:/Users/gabia/Desktop/python_homework_gabi/data/advent_of_code_3.1.txt")
```

```python
def max_joltage(s): 
    max_sofar_1 = 0
    max_index_1 = 0
    
    #first max:
    for i, x in enumerate(s[:-1]): 
        if int(x) > max_sofar_1: 
            max_sofar_1 = int(x)
            max_index_1 = i  

    # second max: 
    max_sofar_2 = 0
    for i, x in enumerate(s[max_index_1 + 1:]): 
        if int(x) > max_sofar_2: 
            max_sofar_2 = int(x) 
    
    # give combined number of both max:
    return int(str(max_sofar_1)+(str(max_sofar_2)))
```

```python
final_numbers = []
number = 0 

for x in data: 
    number = max_joltage(x)
    final_numbers.append(number)
print(sum(final_numbers))
```

```python
def priority(c):
    if c.islower():
        return ord(c) - ord('a') + 1
    else:
        return ord(c) - ord('A') + 27

# Group data in sets of 3 lines
final_numbers_part2 = []
for i in range(0, len(data), 3):
    group = data[i:i+3]
    if len(group) == 3:
        # sets of letters
        sets = [set(line) for line in group]
        # intersection â†’ common letter
        common = sets[0].intersection(sets[1], sets[2])
        if common:
            # take the first common letter and convert to priority
            final_numbers_part2.append(priority(list(common)[0]))

print(sum(final_numbers_part2))
```
