```python
# take in txt file & converting every line into list:

def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
#defining function to get new position after dialing

def rotate_dial(position, rotation):
    # seperating instructions:
    direction = rotation[0]
    steps = int(rotation[1:])
    
    # movement direction:
    if direction == 'L':
        position = position - steps 
    else: # for R 
        position = position + steps
                    
    # adjusting to respect boundary of [0-99]:
    if position < 0: 
        position += 100
    elif position > 99:
        position -= 100 

    position = position % 100     # for when number is way out of range
    
    return(position)
```

```python
def rotate_and_count_passes(position, rotation):
    direction = rotation[0]
    steps = int(rotation[1:])
    
    zero_passes = 0
    
    for _ in range(steps):
        if direction == 'L':
            position -= 1
        else:  # R
            position += 1
            
        position = position % 100
        
        if position == 0:
            zero_passes += 1
            
    return position, zero_passes
```

```python
position = 50
second_password = 0

# converting txt file into instructions list
instructions = read_file_into_lines("C:/Users/gabia/Desktop/python_homework_gabi/data/advent_of_code_1.1.txt")


for rotation in instructions:
    position, passes = rotate_and_count_passes(position, rotation)
    second_password += passes

print(second_password)
```
