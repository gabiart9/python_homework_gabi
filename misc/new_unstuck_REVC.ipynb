```python
# making string of reverse strand into list
base_seperate = list(result)
print(base_seperate)
```

```python
# complementing reverse strand

addition =[]

for base in base_seperate: 
    if base == 'T': 
        addition.append('A')
    elif base == 'A':
        addition.append('T')
    elif base == 'G':
        addition.append('C')
    elif base == 'C':
        addition.append('G')

print(addition)
```
