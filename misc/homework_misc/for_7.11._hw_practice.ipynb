When comparing characters (like 'a', 'A', '0'), Python uses the Unicode (or ASCII) order — that’s basically how characters are numbered internally by the computer.
The order is roughly:

0–9  <  A–Z  <  a–z

So numerically:

'0' → 48

'A' → 65

'a' → 97

Since min() picks the smallest value, '0' wins.

```python
print(min("b", "z", "l", "o"))
```

Here are a few examples:

Character	Code point (number)
'0'	48
'A'	65
'B'	66
'a'	97
'b'	98
'i'	105

When you compare letters (like 'b' and 'i') using min() or max(), Python checks their numeric code values — not how they look to us.

So:

'b' → 98
'i' → 105


Since 98 < 105 → 'b' is smaller than 'i'

```python
my_string = 'Hello world!'  # creation of a string object 

print(len(my_string))       # the len function takes a string as an argument and returns the length of the string

print(my_string.swapcase()) # calling the swapcase method on the my_string object

print(my_string.__len__())  # calling the internal __len__ method on the my_string object, used by len(my_string)
```

```python
print(my_string.isupper())          # Not all the letters are uppercase
print(my_string.upper())            # This capitalizes all the letters

print(my_string.upper().isupper())  # Now all the letters are uppercase
```

Calculate the GC content of the following sequence:

```python
dna = "CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGACTGGGAACCTGCGGGCAGTAGGTGGAAT"
gc_count = dna.count('G') + dna.count('C')
gc_content = (gc_count / len(dna)) * 100
print(f"GC Content: {gc_content:.2f}%")
```

explanaition:
dna.count('G') → counts how many G bases are in the string
dna.count('C') → counts how many C bases are in the string
Add them together to get total G + C count
Divide by total sequence length (len(dna))
Multiply by 100 to get a percentage

```python
radiance = 1.0
radiance = max(2.1, 2.0 + min(radiance, 1.1 * radiance - 0.5))
```

```python
print(radiance)
```

Order of operations:
1.1 * radiance = 1.1
1.1 - 0.5 = 0.6
min(radiance, 0.6) = 0.6
2.0 + 0.6 = 2.6
max(2.1, 2.6) = 2.6
At the end, radiance = 2.6
