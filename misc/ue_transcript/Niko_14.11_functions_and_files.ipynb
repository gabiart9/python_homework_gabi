### Initializing challenge

Python allows you to intialize values in groups:

```python
a, b, c = 0, 3, 5
print(a, b, c)
# is equivalent to
a = 0
b = 3
c = 5
print(a, b, c)
```

```python
values = [1,1,1,1,1,1]
smallest = None
largest = None
for v in values:
    if smallest is None and largest is None:
        smallest, largest = v, v
    else:
        smallest = min(smallest, v)
        largest = max(largest, v)
print(smallest, largest)
```

### reading files into a list of lines

we will need to read lots of files for Rosalind and advent of code. Here's how i would do it:

```python
file_path = "/Users/npapadop/Downloads/rosalind_ini6.txt"
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
line_list = read_file_into_lines(file_path)
```
