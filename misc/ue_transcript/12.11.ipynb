### positions in list:

# Rosalind: INI4

Given: Two positive integers a and b (a<b<10000).  
Return: The sum of all odd integers from a through b, inclusively.

a + b included  
usually in [a, b] b not included in python !!

odd number is not dividable by 2 so:   
odd = 2*n +1  
because 1 is left over after division by 2 we can look for 1 after that  
% = left over after division  

```python
# given a list of numbers, go over them + check if they are odd

numbers = [1,2,3,4,5,6,7]

for n in numbers: # loop over the numbers
    if n % 2 == 1: # check if it is odd 
        print(n)
```

```python
numbers = [1,2,3,4,5,6,7]
result = 0

for n in numbers: 
    if n % 2 == 1: 
        result = result + n
print(result)
```

```python
words = ["and", "and", "test", "kiwi"]

frequencies = {
    "and" : 2,
    "test" : 1, 
    "kiwi" : 1
}
```

```python
# given one word, how would we find its frequency ? 

query ='kiwi'
count = 0
for word in words:
    if word == query:
        count = count + 1
print('the word', query, 'shows up', count, 'times')
```

```python
words = ["and", "and", "test", "kiwi"]

# given one word, how would we find its frequency ?
search = "and"

frequencies = {}

frequencies[search] = 0 

for word in words:
    if word == search:
        frequencies[search] = frequencies[search] + 1

frequencies
```
