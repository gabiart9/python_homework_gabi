```python
# formula for calculating GC-content
# returning multiplied by 100 to get % not 0.x

seq = 'AAACCGGTTT'
def gc_content(seq):
    total_Gs = seq.count('G')
    total_Cs = seq.count('C')
    percent_gc = (total_Gs + total_Cs) / len(seq)
    return 100 * percent_gc
print(gc_content(seq))
```
