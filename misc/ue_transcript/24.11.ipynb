```python
def rabbits(n, k = 1):
    if n == 1:
        return 1
    if n == 2: 
        return 1 
    if n == 3: 
        return 1 + k

    return rabbits (n-1, k) + rabbits (n-2, k)
```

```python
n = 5 
k = 3 
rabbits(n, k)
```

```python
def rabbits(n, k = 1):
    if n == 1:
        return 1
    if n == 2: 
        return 1 

    return rabbits (n-1, k) + rabbits (n-2, k)*k
```
