# Rosalind DNA:

## DNA original

```python
s = "CATCGGCGACCTCT"

count_A = s.count("A")
count_C = s.count("C")
count_G = s.count("G")
count_T = s.count("T")

print(count_A, count_C, count_G, count_T)
```

## DNA with dictionaries

```python
DNA = "CATCGGCGACCTCT"
count = {
    "A": DNA.count("A"),
    "C": DNA.count("C"),
    "G": DNA.count("G"),
    "T": DNA.count("T"),
}
print(count)
```

## DNA with loops

```python
loop_DNA = "CATCGA"

letters = loop_DNA.split()
print(letters)

frequencies = {}
for base in letters: 
    frequencies[base] = 0

for base in letters: 
    frequencies[base] = frequencies[base] + 1

for key, value in frequencies.items():
    print(key, value)
```
