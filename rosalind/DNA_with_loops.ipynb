# Rosalind DNA:

## DNA original

```python
s = "CATCGGCGACCTCT"

count_A = s.count("A")
count_C = s.count("C")
count_G = s.count("G")
count_T = s.count("T")

print(count_A, count_C, count_G, count_T)
```

## DNA with dictionaries

```python
DNA = "CATCGGCGACCTCT"
count = {
    "A": DNA.count("A"),
    "C": DNA.count("C"),
    "G": DNA.count("G"),
    "T": DNA.count("T"),
}
print(count)
```
