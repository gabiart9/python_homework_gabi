# GC

```python
def gc_content(seq):
    if len(seq) == 0:
        return 0 
    total_Gs = seq.count('G')
    total_Cs = seq.count('C')
    percent_gc = (total_Gs + total_Cs) / len(seq)
    return 100 * percent_gc
```

```python
current_sequence = ''
current_id = ''
sequences = {}

with open("C:/Users/gabia/Desktop/python_homework_gabi/data/rosalind_gc (1).txt") as fasta:
    for line in fasta:
        if line[0] == '>':   # same as: "if line.startswith('>'):
            if current_id is not None:
                sequences[current_id] = current_sequence
            current_id = line[1:]    # store new id
            current_sequence = ''    # reset sequence to not get overwritten
        else:
            current_sequence = current_sequence + line        # append sequences

if current_id is not None:
    sequences[current_id] = current_sequence
```

```python
maximum_gc = -1 
maximum_gc_id = ''

for seq_id, seq in sequences.items(): 
    gc = gc_content(seq)
    # is this the highest GC content we've seen ? 
    if gc > maximum_gc: 
        maximum_gc = gc
        maximum_gc_id = seq_id

print(maximum_gc_id)
print(maximum_gc)
```
