Given: A string s of length at most 10000 letters.

Return: The number of occurrences of each word in s, where words are separated by spaces.  
Words are case-sensitive, and the lines in the output can be in any order.

```python
s = "Apple bottom jeans"
```

```python
words = s.split
print(words)
```

```python
# given one word, how would we find its frequency ?
query = "and"
```

```python
frequencies = {}

frequencies[query] = 0 

for word in words:
    if word == query:
        frequencies[query] = frequencies[query] + 1

frequencies
```
