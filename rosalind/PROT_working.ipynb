```python
sample_dataset = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
counter = 0 # count the position of the beginning and end of the codons in the string 
aa = ""
for codon in sample_dataset:
    triplet = sample_dataset[counter:counter + 3] 
    if triplet == '': # at the end of a string we want to exit the loop
        exit
    else:
        aa = aa + codon_sun[triplet] 
        counter = counter + 3
print(aa)
```
