# Mendel's First Law

```python
Pr1 = (2/6 * 1/5 * 1) + (2/6 * 2/5 * 1) + (2/6 * 2/5 * 1)
print(Pr1)
```

```python
Pr2 = (2/6 * 2/5 * 1) + (2/6 * 1/5 *3/4) + (2/6 * 2/5 * 1/2)
print(Pr2) 
```

```python
Pr3 = (2/6 * 2/5 * 1) + (2/6 * 2/5 * 1/2)
print(Pr3)
```

```python
PR = Pr1 + Pr2 + Pr3  
print(PR)
```

```python
k = 16
m = 28
n = 18
N = k + m + n 

homozygot_dominant = (k/N * (k-1)/(N-1) * 1) + (k/N * m/(N-1) * 1) + (k/N * n/(N-1) * 1)
heterozygot = (m/N * k/(N-1) * 1) + (m/N * (m-1)/(N-1) * 3/4) + (m/N * n/(N-1) * 1/2)
homozygot_rezessiv = (n/N * k/(N-1) * 1) + (n/N * m/(N-1) * 1/2) 
print(homozygot_dominant, heterozygot, homozygot_rezessiv)
Answer = homozygot_dominant + heterozygot + homozygot_rezessiv 
print(f"Answer is: {Answer}") 
```
